```python
import pandas as pd
df = pd.read_parquet("steam.parquet")
print("Veri boyutu: ", df.shape)
print("Sütunlar: \n",df.columns.to_list())
print(df.head(3))
```

```python
# Eksik veri kontrolü
df.info()
```

```python
# İlk kontorl
print("Eksik değerler: \n")
df.isna().sum().sort_values(ascending=False).head(15)
```

```python
df = df.drop(columns=["Score rank", "Metacritic url", "Reviews", "Notes", "Website", "Support url", "Support email", "Header image", "Movies", "Screenshots", "About the game"])
```

```python
# 2. kontrol
df.isna().sum().sort_values(ascending=False).head(7)
```

```python
# Kategorik sütunlardaki boş veirleri uygun olarak dolduralım
df["Tags"] = df["Tags"].fillna("Unknown")
df["Categories"] = df["Categories"].fillna("Unknown") 
df["Publishers"] = df["Publishers"].fillna("Unknown Publisher")
df["Developers"] = df["Developers"].fillna("Unknown Developer")
df["Genres"] = df["Genres"].fillna("Unknown")

# Boş "Name" olan oyunları silelim
df = df.dropna(subset=["Name"])
```

```python
# 3. kontrol
df.isna().sum().sort_values(ascending=False)
```
